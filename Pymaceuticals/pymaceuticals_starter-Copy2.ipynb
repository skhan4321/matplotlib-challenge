```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os 

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_file = "data/mouse_drug_data.csv"
clinical_trial_file = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_df=pd.read_csv(mouse_file)
clinical_df=pd.read_csv(clinical_trial_file)


# Combine the data into a single dataset
dataframe=pd.merge(mouse_df,clinical_df, on="Mouse ID",how="outer")
                   
#http://www.datasciencemadesimple.com/join-merge-data-frames-pandas-python/)


# Display the data table for preview
dataframe.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

mtv=dataframe.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean()



# Convert to DataFrame
mtvdf=pd.DataFrame(mtv)
mtv_new_df=mtv.reset_index() 
# Preview DataFrame
mtv_new_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stderror=dataframe.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem() 
stderror_ri=stderror.reset_index()
# Convert to DataFrame
stderror_df=pd.DataFrame(stderror_ri)
# Preview DataFrame
stderror_df.head()
```

```python
df_reformated_sem=stderror_df.pivot(index="Timepoint", columns="Drug",values="Tumor Volume (mm3)")
df_reformated_sem
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_reformated=mtv_new_df.pivot(index="Timepoint", columns="Drug",values="Tumor Volume (mm3)")

# Preview that Reformatting worked
df_reformated
```

```python
#plt.errorbar(df_reformated.index - this is the x since its also the index),

#df.reformated[Drug]) - this is the y axis and include bracket because its the column name 
```

```python
# Generate the Plot (with Error Bars)


plt.errorbar(df_reformated.index,df_reformated["Capomulin"],marker='v', linewidth=1.0, linestyle="solid", yerr=df_reformated_sem["Capomulin"])
plt.errorbar(df_reformated.index,df_reformated["Infubinol"],marker='p', linewidth=1.0, linestyle="solid", yerr=df_reformated_sem["Infubinol"])
plt.errorbar(df_reformated.index,df_reformated["Ketapril"],marker='s', linewidth=1.0, linestyle="solid", yerr=df_reformated_sem["Ketapril"])
plt.errorbar(df_reformated.index,df_reformated["Placebo"],marker='x', linewidth=1.0, linestyle="solid", yerr=df_reformated_sem["Placebo"])


plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(["Capomulin","Infubinol","Ketapril","Placebo"])


# Save the Figure
plt.show()
```
