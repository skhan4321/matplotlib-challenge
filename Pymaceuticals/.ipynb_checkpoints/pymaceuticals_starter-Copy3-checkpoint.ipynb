```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os 

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_file = "data/mouse_drug_data.csv"
clinical_trial_file = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_df=pd.read_csv(mouse_file)
clinical_df=pd.read_csv(clinical_trial_file)


# Combine the data into a single dataset
dataframe=pd.merge(mouse_df,clinical_df, on="Mouse ID",how="outer")
                   
#http://www.datasciencemadesimple.com/join-merge-data-frames-pandas-python/)


# Display the data table for preview
dataframe.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

mtv=dataframe.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean()



# Convert to DataFrame
mtvdf=pd.DataFrame(mtv)
mtv_new_df=mtv.reset_index() 
# Preview DataFrame
mtv_new_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stderror=dataframe.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem() 
stderror_ri=stderror.reset_index()
# Convert to DataFrame
stderror_df=pd.DataFrame(stderror_ri)
# Preview DataFrame
stderror_df.head()
```

```python
df_reformated_sem=stderror_df.pivot(index="Timepoint", columns="Drug",values="Tumor Volume (mm3)")
df_reformated_sem
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_reformated=mtv_new_df.pivot(index="Timepoint", columns="Drug",values="Tumor Volume (mm3)")

# Preview that Reformatting worked
df_reformated
```

```python
#plt.errorbar(df_reformated.index - this is the x since its also the index),

#df.reformated[Drug]) - this is the y axis and include bracket because its the column name 
```

```python
# Generate the Plot (with Error Bars)


plt.errorbar(df_reformated.index,df_reformated["Capomulin"],marker='v', linewidth=1.0, linestyle="solid", yerr=df_reformated_sem["Capomulin"])
plt.errorbar(df_reformated.index,df_reformated["Infubinol"],marker='p', linewidth=1.0, linestyle="solid", yerr=df_reformated_sem["Infubinol"])
plt.errorbar(df_reformated.index,df_reformated["Ketapril"],marker='s', linewidth=1.0, linestyle="solid", yerr=df_reformated_sem["Ketapril"])
plt.errorbar(df_reformated.index,df_reformated["Placebo"],marker='x', linewidth=1.0, linestyle="solid", yerr=df_reformated_sem["Placebo"])


plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(["Capomulin","Infubinol","Ketapril","Placebo"])


# Save the Figure
plt.savefig('tumor.png')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
stderror_ms = dataframe.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem().reset_index()
# Convert to DataFrame
stderror_ms_df=pd.DataFrame(stderror_ms)
# Preview DataFrame
```

```python
df_reformated_ms = stderror_ms_df.pivot(index="Timepoint", columns="Drug",values='Metastatic Sites')
df_reformated_ms

```

```python
# Minor Data Munging to Re-Format the Data Frames

ms_reformated_sem=mtri.pivot(index="Timepoint", columns="Drug",values="Metastatic Sites")
# Preview that Reformatting worked
ms_reformated_sem

#ms_reformated_sem=stderror_ms_df.pivot(index="Timepoint", columns="Drug",values="Metastatic Sites")
# Preview that Reformatting worked
#ms_reformated_sem


```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(df_reformated_ms.index,ms_reformated_sem["Capomulin"],marker='v', linewidth=1.0, linestyle="solid", yerr=df_reformated_ms["Capomulin"])
plt.errorbar(df_reformated_ms.index,ms_reformated_sem["Infubinol"],marker='p', linewidth=1.0, linestyle="solid", yerr=df_reformated_ms["Infubinol"])
plt.errorbar(df_reformated_ms.index,ms_reformated_sem["Ketapril"],marker='x', linewidth=1.0, linestyle="solid", yerr=df_reformated_ms["Ketapril"])
plt.errorbar(df_reformated_ms.index,ms_reformated_sem["Placebo"],marker='s', linewidth=1.0, linestyle="solid", yerr=df_reformated_ms["Placebo"])


plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastic Spread")
plt.legend(["Capomulin","Infubinol","Ketapril","Placebo"])

# Save the Figure
plt.savefig('metastatic_spread.png')
# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = dataframe.groupby(['Drug', 'Timepoint'])['Mouse ID'].count()

# Convert to DataFrame
mouse_countdf = pd.DataFrame(mouse_count)
mouse_count_ri = mouse_countdf.reset_index()
# Preview DataFrame
mouse_count_ri
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_reformatted = mouse_count_ri.pivot(index='Timepoint', columns='Drug', values='Mouse ID')

# Preview the Data Frame
mouse_count_reformatted
```

```python
survival_rate = mouse_count_reformatted/(mouse_count_reformatted.iloc[0]).round(0)*100
survival_rate
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(mouse_count_reformatted.index,survival_rate["Capomulin"],marker='v', linewidth=1.0, linestyle="solid")
plt.plot(mouse_count_reformatted.index,survival_rate["Infubinol"],marker='p', linewidth=1.0, linestyle="solid")
plt.plot(mouse_count_reformatted.index,survival_rate["Ketapril"],marker='x', linewidth=1.0, linestyle="solid")
plt.plot(mouse_count_reformatted.index,survival_rate["Placebo"],marker='s', linewidth=1.0, linestyle="solid")

plt.title("Survival during Treatmentt")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate(%) ")
plt.legend(["Capomulin","Infubinol","Ketapril","Placebo"])
         
# Save the Figure
plt.savefig('survival.png')
# Show the Figure
plt.show()
```
